## Ch01 추천 시스템 소개

### 1.1 주요 추천 알고리즘

__1. 협업 필터링(Collaborative Filtering : CF)__
+ 구매 또는 소비한 제품에 대한 각 소비자의 평가를 받아서 평가 패턴이 비슷한 소비자를 한 집단으로 보고 그 집단에 속한 소비자들의 취향을 활용하는 기술이다.
+ 소비자들의 평가정보(explicit data)를 얻기 어려운 경우가 많아 클릭 수나 접속 횟수 등 간접적인 데이터(implicit data)를 사용하는 경우가 많다.
+ 예로 아마존의 제품 추천 시스템과 넷플릭스 영화 추천 시스템이 있다.

__2. 내용 기반 필터링(Content-Based filtering : CB)__
+ 제품의 내용을 분석해서 추천하는 기술을 뜻한다.
+ 소비자가 소비하는 제품 중 텍스트 정보가 많은 제품을 분석하여 추천할 때 많이 이용되는 기술이다.

__3. 지식 기반 필터링(Knowledge-Based filtering : KB)__
+ 특정 분야 전문가의 도움을 받아서 그 분야에 대한 전체적인 지식구조를 만들어서 활용하는 방법이다.
+ 지식이 중요한 분야에서 효과가 좋지만 전문가가 필요하고 전문가의 능력에 의존하므로 세세한 분야에 적용하기 어렵다.

__4. 딥러닝 추천 기술(Deep Learning : DL)__
+ 딥러닝의 입력으로 다양한 사용자와 아이템의 특징값을 사용하고, 이를 통해 얻은 출력은 각 아이템에 대한 각 사용자의 예상 선호도로 사용하는 것이 보통이다.
+ 장점으로 다양한 입력 변수를 사용할 수 있어 특히 이미지 인식 등에서 큰 성능을 발휘한다.
+ 연속값(평점)으로 표시되는 아이템에 대한 추천에서는 정확도가 다른 알고리즘에 비해 크게 우월하지는 않다고 한다.

__5. 하이브리드 기술(Hybrid)__
+ 실제로 추천 시스템을 구축할 때는 한 가지만 사용하는 경우는 별로 없고 대부분 두 가지 이상의 기술을 혼합하여 사용하는 하이브리드 형태로 자주 사용된다.
+ 연구에 따르면 하나의 기술을 사용할 때보다 복수의 기술을 사용하는 경우 정확도가 향상된다고 한다.
